Task 2 - Visualise the distributions of ratings and compensations

We will now visualise the distributions of employee ratings and compensations

We will create the following plots:

A scatter plot of employee age and employee salary, healthy_eating and active_lifestyle, health_eating and salary

A count plot of groups

A histogram of employee salary

```python
plt.figure(figsize=(10,5))
plt.scatter(data=data,x='age',y='salary',color='blue',
               edgecolor='linen',alpha=.5)
plt.title("scatterplot of employee age vs employee salary")
plt.xlabel("employee age")
plt.ylabel("employee salary")
plt.show()
```

we observe that in general there is a relationship between age and salary of employee
as age increases the salary also increases though there are some outliers present as well

```python
plt.figure(figsize=(10,5))
plt.scatter(data=data,x='healthy_eating',y='active_lifestyle',color='green',
               edgecolor='linen',alpha=.5)
plt.title("scatterplot of healthy_eating vs active_lifestyle")
plt.xlabel(" healthy_eating")
plt.ylabel("active_lifestyle")
plt.show()
```

we observe that in general there is a relationship b/t healthy_eating and active eating of employee
as healthy eating increases the active lifestyle also increases

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=data,x='healthy_eating',y='salary',color='orange',
               edgecolor='linen',alpha=.5)
plt.title("scatterplot of healthy_eating vs salary")
plt.xlabel(" healthy_eating")
plt.ylabel("salary")
plt.show()
```

we observe that in general there is a relationship b/t healthy_eating and salary of employee as healthy eating increases the salary also increases

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data,x="groups",
               edgecolor='linen',alpha=.7)
plt.title("countplot of groups")
plt.xlabel(" groups")
plt.ylabel("count")
plt.show()
```

we see that most employees either belong to group a or group O with group A having  maximum frequency

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x='salary',color='orange',edgecolor='linen',alpha=.5,bins=10)
plt.title("histogram of salary")
plt.xlabel("salary")
plt.ylabel("count")
plt.show()
```

Observations:

We see that the salaries are not uniform, which is not necessarily a discrepancy But we do observe that there are employees with salaries on either extremities of the histogram Based on the histogram, we can see that the count of highest earning employee is very less

Task 3 - Subset the data based on thresholds

We will now subset the original data frame based on the following conditions:

Employees with healthy_eating greater than 8

Employees with Salary less than 1000

Employees with healthy_eating greater than 8 and with salary less than 1000

```python
sub1=data[data['healthy_eating']>8]
sub1
```

```python
#salary less tthan 1000
sub2=data[data['salary']<1000]
sub2
```

```python
#employees with healthy_eating greater than 8 and with salary less than 1000
sub=data[(data['healthy_eating']>8) & (data['salary']<1000)]
sub
```
